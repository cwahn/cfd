```python
from pathlib import Path

from numpy import ndarray


def get_points_from_dataset(file_name: Path) -> ndarray:
    with h5py.File(file_name, "r+") as f:
        data = f.get("points")[:]
        return data 
```

```python
from typing import List
import os


def get_subdir_paths(dir_name: Path) -> List[Path]:
    subdir_names = os.listdir(dir_name)
    subdir_paths = list(map(lambda x: os.path.join(dir_name, x), subdir_names))
    return subdir_paths
```

```python
get_subdir_paths(ROOT_OBJ_DIR)
```

```python
get_subdir_paths('./data/yeouido/visual_feature_map/1.0.0/15/24026')
```

```python
from functools import reduce
from operator import iconcat


def get_dataset_paths(base_obj_dir: Path) -> List[Path]:
    obj_paths = get_subdir_paths(base_obj_dir)
    dataset_pathss = list(map(lambda x: get_subdir_paths(x), obj_paths))
    dataset_paths = list(reduce(iconcat, dataset_pathss, []))
    return dataset_paths
```

```python
get_dataset_paths(ROOT_OBJ_DIR)
```

```python
from numpy import concatenate
import numpy as np


def get_points(root_obj_dir: Path) -> ndarray:
    dataset_paths = get_dataset_paths(root_obj_dir)
    pointss = list(map(lambda x: get_points_from_dataset(x), dataset_paths))
    pointss = list(map(np.ndarray.tolist, pointss))
    points = list(reduce(iconcat, pointss, []))
    return points
```

```python
points = get_points(ROOT_OBJ_DIR)
```
